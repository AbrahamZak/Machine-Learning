# Linear Least Square Regression

## Import neccessary libraries


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Pull data from our data file
Read the data file and splits the columns into x and y, then processes for each degree

```python
if __name__ == '__main__':
    #read text file
    x = np.genfromtxt(fname="hw2.dat", usecols=0)
    y = np.genfromtxt(fname="hw2.dat", usecols=1)
```

## Polynomial (Degree: 1)
Gets the polynomial from x, y, and the degree of 1\
Output the parameters of the polynomial and the fitting error

```python
    z = np.polyfit(x, y, 1)
    print("Parameters of polynomial of degree 1:", z)
    print ("Fitting error: ", np.sum((np.polyval(z, x) - y)**2))
```

## Results/Plot (Degree: 1)
Generates a graph for the degree with the original data

```python
    plt.title("Linear least square regression degree: 1")
    plt.xlabel("x")
    plt.ylabel("y")
    #plot original data
    plt.plot(x,y, '.', label='data')
    #plot the fit 
    plt.plot(x,np.polyval(z,x), '-', label='regression')
    plt.legend();
    plt.show()
```
